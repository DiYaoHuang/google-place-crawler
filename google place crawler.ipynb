```python
from googlemaps.client import Client
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import time
import random
```

```python
def getListItem(list):
    res = ''
    for r in list:
        res += r + ','
    res = res[:-1]
    return res
def getTag(tag):
    tags = []
    for i in range(len(tag)):
        tags.append(tag[i][1])
    return  tags
def getServiceTypes(service_type):
    service_types = []
    for i in range(len(service_type)):
        service_types.append(service_type[i][1])
    return  service_types
def getBusyPeriod(busy_period):
    busy_period_by_weekday = {}
    for i in range(len(busy_period)):
        busy_period_by_day_dict = {}
        if(i + 7 == busy_period[0][0]):
            for r in range(len(busy_period[0][1])):
                interval_hour = {busy_period[0][1][r][0] : busy_period[0][1][r][1]}
                busy_period_by_day_dict.update(interval_hour)
            busy_period_by_weekday_dic = {"星期日" : sorted(busy_period_by_day_dict, key=busy_period_by_day_dict.get, reverse=True)[:5]}
            busy_period_by_weekday.update(busy_period_by_weekday_dic)
        elif(i == busy_period[1][0]):
            for r in range(len(busy_period[0][1])):
                interval_hour = {busy_period[0][1][r][0] : busy_period[0][1][r][1]}
                busy_period_by_day_dict.update(interval_hour)
            busy_period_by_weekday_dic = {"星期一" : sorted(busy_period_by_day_dict, key=busy_period_by_day_dict.get, reverse=True)[:5]}
            busy_period_by_weekday.update(busy_period_by_weekday_dic)
        elif(i == busy_period[2][0]):
            for r in range(len(busy_period[0][1])):
                interval_hour = {busy_period[0][1][r][0] : busy_period[0][1][r][1]}
                busy_period_by_day_dict.update(interval_hour)
            busy_period_by_weekday_dic = {"星期二" : sorted(busy_period_by_day_dict, key=busy_period_by_day_dict.get, reverse=True)[:5]}
            busy_period_by_weekday.update(busy_period_by_weekday_dic)
        elif(i == busy_period[3][0]):
            for r in range(len(busy_period[0][1])):
                interval_hour = {busy_period[0][1][r][0] : busy_period[0][1][r][1]}
                busy_period_by_day_dict.update(interval_hour)
            busy_period_by_weekday_dic = {"星期三" : sorted(busy_period_by_day_dict, key=busy_period_by_day_dict.get, reverse=True)[:5]}
            busy_period_by_weekday.update(busy_period_by_weekday_dic)
        elif(i == busy_period[4][0]):
            for r in range(len(busy_period[0][1])):
                interval_hour = {busy_period[0][1][r][0] : busy_period[0][1][r][1]}
                busy_period_by_day_dict.update(interval_hour)
            busy_period_by_weekday_dic = {"星期四" : sorted(busy_period_by_day_dict, key=busy_period_by_day_dict.get, reverse=True)[:5]}
            busy_period_by_weekday.update(busy_period_by_weekday_dic)
        elif(i == busy_period[5][0]):
            for r in range(len(busy_period[0][1])):
                interval_hour = {busy_period[0][1][r][0] : busy_period[0][1][r][1]}
                busy_period_by_day_dict.update(interval_hour)
            busy_period_by_weekday_dic = {"星期五" : sorted(busy_period_by_day_dict, key=busy_period_by_day_dict.get, reverse=True)[:5]}
            busy_period_by_weekday.update(busy_period_by_weekday_dic)
        elif(i == busy_period[6][0]):
            for r in range(len(busy_period[0][1])):
                interval_hour = {busy_period[0][1][r][0] : busy_period[0][1][r][1]}
                busy_period_by_day_dict.update(interval_hour)
            busy_period_by_weekday_dic = {"星期六" : sorted(busy_period_by_day_dict, key=busy_period_by_day_dict.get, reverse=True)[:5]}
            busy_period_by_weekday.update(busy_period_by_weekday_dic)
    return busy_period_by_weekday
def getRecommend(recommend):
    recommends = ''
    for i in range(len(recommend)):
        try:
            recommend[i][1][11]
            recommends += recommend[i][1][11] + ';' + getListItem(recommend[i][1][13]) + ';'
        except:
            recommends += ''
    return recommends
def getreview(phase, review):
    p = 0 if phase == 1 else 2
    reviews = {}
    for i in range(len(review[p])):
        review_item = []
        user_ID = review[p][i][6]
        user_name = review[p][i][0][1]
        time = review[p][i][1]
        rate = review[p][i][4]
        likes = review[p][i][16]   
        content = review[p][i][3] if review[p][i][3] is not None else ''     
        review_item = [user_ID, user_name, time, rate, content, likes]
        reviews.update({user_ID : review_item})
    return reviews
```

```python
#ChIJ714HdQmpQjQRn0TTQzMCCbA 屈臣氏Watsons 西門店
#ChIJhfQwjL2rQjQR9h4iphLNidw Toyata松山服務廠
#ChIJzzNFLVmpQjQR-vuhAtSqMkQ 唐宮蒙古烤肉涮羊肉餐廳
#ChIJh-J5vhsRbjQR4AcXKxaTRt8 義大遊樂世界
placeId = 'ChIJObHbu4wjaDQRhUgtv_paoSo'
info = gmaps.place(place_id = placeId, language='zh-TW')
url = info.get('result').get('url')
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
scripts = soup.head.find_all('script')
for script in scripts:
    if 'window.APP_INITIALIZATION_STATE' in script.contents[0]:
        script_text = script.contents[0]
ds_text = script_text[script_text.index('window.APP_INITIALIZATION_STATE')+32:script_text.index(';window.APP_FLAGS')]
ds_text = ds_text.replace(")]}\'", '')
ds = json.loads(ds_text)
ds2 = json.loads(ds[3][6])
```
